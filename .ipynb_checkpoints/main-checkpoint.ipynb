```python
import pandas as pd
import numpy as np
import matplotlib as mpl
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import OneHotEncoder, MinMaxScaler, StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
#загружаем данные
train = pd.read_csv('train.csv').drop(columns=['Unnamed: 0'])
test = pd.read_csv('test.csv').drop(columns=['Unnamed: 0'])
```

```python
#соединим трейн и тест, чтобы извлечь все номера гейтов
all_data = pd.concat([train, test])
gate_ids = sorted(all_data['gate_id'].unique())
```

```python
def little_pipeline(df: pd.DataFrame) -> pd.DataFrame:
    '''
    Выполняет основную предподготовку данных: получает день, месяц, час, минуту, секунду, день недели, час+минуты в виде float, номер декады месяца.
    '''
    df['ts'] = pd.to_datetime(df['ts'])
    df['day'] = df['ts'].dt.day
    df['month'] = df['ts'].dt.month
    df['hour'] = df['ts'].dt.hour
    df['minutes'] = df['ts'].dt.minute
    df['seconds'] = df['ts'].dt.second
    df['weekday'] = df['ts'].dt.weekday
    df['hour_minutes'] = df['ts'].dt.minute / 60 + df['ts'].dt.hour
    df['decade'] = df['day'].apply(lambda x: month_decade(x))

    return df.sort_values(by='ts')

def month_decade(day: int) -> str:
    '''
    Получает категориальный номер декады месяца.
    '''
    if 0 < day <= 10:
        return 'decade_1'
    elif 10 < day < 21:
        return 'decade_2'
    else:
        return 'decade_3'

def clean_errors(df:pd.DataFrame, column:str) -> pd.DataFrame:
    '''
    Очищает данные от дубликатов.
    В системе контроля и управления доступом, видимо, при проходе через турникет, иногда происходит задвоение считывания пропуска. 
    Поэтому мы убираем дублицированные проходы через один и тот же турникет, выполненные одним и тем же человеком в течении минуты.
    '''
    a = df.groupby([
        column,
        'gate_id',
        'month',
        'day',
        'hour',
        'minutes',
        #'seconds'
        ]).agg({'ts':['first','last', 'count']}).reset_index()
    
    a.columns = [column,
                'gate_id',      
                  'month',      
                    'day',      
                   'hour',      
                'minutes',      
               # 'seconds',      
                    'first',
                    'last',
                     'count']
    
    df = df.merge(how='right',
    right=a[[column, 'first']],
    left_on=[column, 'ts'],
    right_on=[column, 'first']).drop(columns=['first'])

    df = df.drop_duplicates()
    return df.sort_values(by='ts')

def create_grouped_by_day(df:pd.DataFrame, column:str) -> pd.DataFrame:
    '''
    Создание сводной таблицы для извлечения информации по проходам пользователей за день.
    '''
    df_grouped = df.sort_values(by='ts').groupby([column, 'month', 'day', 'weekday']).agg({'gate_id':['first', 'last', 'unique'],
                                                        'ts':['min','max'],
                                                       'hour':['min', 'max', 'unique'],
                                                       'hour_minutes':['min', 'max'],
                                                        'day':['count']}
                                                      )
    return df_grouped

def time_after_previous(df:pd.DataFrame, grouped_df:pd.DataFrame, column:str) -> pd.DataFrame:
    '''
    Создает колонки и добавляет в результирующую таблицу с ифнормацией о:
    1. Факте того, что данный турникет является первым за день для пользователя, то есть турникетом входа
    2. Факте того. что данный турникет является последним за день для пользователя, то есть турникетом выхода
    3. Времени после предыдущего турникета
    4. Номере предыдущего турникета
    5. Номере следующего турникета
    6. Времени с момента первого захода в данный день
    '''
    result_df = pd.DataFrame(columns = list(df.columns)+['is_enter', 'is_exit' 'time_from_previous',
                                                         'previous_gate', 'next_gate', 'time_at_job'],
                             index=df.index)
    queries = []
    indexes = grouped_df.index
    for index in indexes:
        user, month, day, weekday = index
        
        grouped_row = grouped_df.loc[grouped_df.index == index]
        query = df[(df[column] == user)  
                   & (df['month'] == month) 
                   & (df['day'] == day) 
                   & (df['weekday'] == weekday)]
        indexes_from_query = query.index
        query = query.reset_index(drop=True)
        for i, row in query.iterrows():
            if row['hour_minutes'] == grouped_row[('hour_minutes', 'min')].values[0]:
                i0 = i
                query.loc[i, 'is_enter'] = 1
                query.loc[i, 'previous_gate'] = -999
                try:
                    query.loc[i, 'next_gate'] = query.loc[i+1, 'gate_id']
                except:
                    query.loc[i, 'next_gate'] = -999
                query.loc[i, 'time_from_previous'] = 0
                query.loc[i, 'time_at_job'] = 0
            elif row['hour_minutes'] == grouped_row[('hour_minutes', 'max')].values[0]:
                query.loc[i, 'is_exit'] = 1
                query.loc[i, 'previous_gate'] = query.loc[i-1, 'gate_id']
                query.loc[i, 'next_gate'] = -999
                query.loc[i, 'time_from_previous'] = query.loc[i, 'hour_minutes'] - query.loc[i-1, 'hour_minutes']
                query.loc[i, 'time_at_job'] = query.loc[i, 'hour_minutes'] - query.loc[i0, 'hour_minutes']
            else:
                query.loc[i, 'not_edge'] = 1
                query.loc[i, 'previous_gate'] = query.loc[i-1, 'gate_id']
                query.loc[i, 'next_gate'] = query.loc[i+1, 'gate_id']
                query.loc[i, 'time_from_previous'] = query.loc[i, 'hour_minutes'] - query.loc[i-1, 'hour_minutes']
                query.loc[i, 'time_at_job'] = query.loc[i, 'hour_minutes'] - query.loc[i0, 'hour_minutes']

        query.index = indexes_from_query
        queries.append(query)
    result_df = pd.concat(queries)
    return result_df.fillna(0)

def get_dummies_gates_moving(df:pd.DataFrame, word:str, gate_ids: list=gate_ids) -> pd.DataFrame:
    '''
    Вспомогательная функция, создающее onehotencoded колонки с номерами гейтов.
    '''
    temp_df = pd.DataFrame(columns = [f'{word}_{gate}' for gate in gate_ids], index = df.index)
    for index, row in df.iterrows():
        temp_df.loc[index][f'{word}_{int(row[word])}'] = 1
    return temp_df.fillna(0)
```

```python
from typing import Tuple
```

```python
def pipeline(df:pd.DataFrame, test:bool=False) -> Tuple[pd.DataFrame]:
    '''
    Функция предподготовки данных для обучения и теста.
    Создает onehotencoded колонки с информацией о:
    1. Факте того, что данный турникет является первым за день для пользователя, то есть турникетом входа
    2. Факте того. что данный турникет является последним за день для пользователя, то есть турникетом выхода
    3. Номере предыдущего турникета
    4. Номере следующего турникета
    5. Номере декады
    6. Дне недели
    Создает standard scaled колонки с информацей о 
    1. Текущем времени в формате дробное число часов (т.е. 10:30 == 10.5).
    2. Времени с предыдущего прохода гейта в формате дробное число часов (т.е. 10:30 == 10.5).
    '''
    if test:
        column = 'user_word'
    else:
        column = 'user_id'
    df = little_pipeline(df)
    df = clean_errors(df, column)
    
    df_grouped = create_grouped_by_day(df, column)
    
    df_d = time_after_previous(df, df_grouped, column)
    
    current_gate = get_dummies_gates_moving(df_d, 'gate_id')
    prev_df = get_dummies_gates_moving(df_d, 'previous_gate')
    next_df = get_dummies_gates_moving(df_d, 'next_gate')
    
    df_decades = pd.get_dummies(df_d['decade']).astype(float)
    df_weekday = pd.get_dummies(df_d['weekday'], prefix='weekday_').astype(float)

    scaler = StandardScaler()
    scaler.fit(np.array(df_d['hour_minutes']).reshape(-1, 1))
    hour_minutes =  pd.DataFrame(columns=['hour_mins'], index = df_d.index, data = scaler.transform(np.array(df_d['hour_minutes']).reshape(-1, 1)))

    scaler = StandardScaler()
    scaler.fit(np.array(df_d['time_from_previous']).reshape(-1, 1))
    time_from_previous =  pd.DataFrame(columns=['time_from_prev'], index = df_d.index, data = scaler.transform(np.array(df_d['time_from_previous']).reshape(-1, 1)))

    df_d = df_d.drop(columns = ['ts', 'month', 'day', 'minutes', 'hour', 'seconds', 'hour_minutes', 'time_from_previous'])

    df_d = pd.concat([df_d[[column, 'is_enter', 'is_exit', 'not_edge']],
                      current_gate,
                      prev_df,
                      next_df,
                      df_decades,
                      df_weekday,
                      hour_minutes,
                      time_from_previous,
          ], axis = 1
         )
    df_d[['is_enter', 'is_exit', 'not_edge']] = df_d[['is_enter', 'is_exit', 'not_edge']].astype(float)
    return df_d
```

```python
train_data = pipeline(train).sample(frac=1).reset_index(drop=True)
test_data = pipeline(test, test=True).sample(frac=1).reset_index(drop=True)
```
